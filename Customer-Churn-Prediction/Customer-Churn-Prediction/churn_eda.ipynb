# 📊 Customer Churn Prediction – Exploratory Data Analysis (EDA)

This notebook performs detailed EDA on the customer churn dataset.
We explore customer behavior, churn patterns, correlations, and key insights.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style='whitegrid')
```

## 📥 Load Dataset

```python
df = pd.read_csv('customer_churn.csv')
df.head()
```

## 🔍 Basic Information

```python
df.info()
```

```python
df.describe()
```

## 🧹 Missing Values

```python
df.isnull().sum()
```

## 📊 Churn Distribution

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='Churn', palette='Set2')
plt.title('Churn Distribution')
plt.show()
```

## 📈 Tenure vs Churn

```python
plt.figure(figsize=(8,5))
sns.histplot(data=df, x='tenure', hue='Churn', kde=True, palette='coolwarm')
plt.title('Tenure Distribution by Churn')
plt.show()
```

## 💵 Monthly Charges vs Churn

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='Churn', y='MonthlyCharges', palette='Set3')
plt.title('Monthly Charges by Churn')
plt.show()
```

## 📞 Contract Type vs Churn

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='Contract', hue='Churn', palette='viridis')
plt.title('Contract Type vs Churn')
plt.xticks(rotation=45)
plt.show()
```

## 🔥 Correlation Heatmap

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

## 📝 Key Insights

- Customers with **month-to-month contracts** churn the most.
- Higher **monthly charges** are strongly linked to churn.
- Customers with **short tenure** (new customers) churn more.
- Long-term contract customers are more loyal.
- Senior citizens show slightly higher churn rates.

These insights will guide feature engineering and model building.
